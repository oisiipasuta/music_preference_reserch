```python
#ヒストグラムでボタン押下位置を確認
music_list = df['song_id'].unique()
music_list = set(music_list)
for music in music_list:
  sns.histplot(data=df[df['song_id'] == music], x='button_pressed_time', binwidth=3)
  plt.show()
```

```python
#各曲の押下率を計算する
for music in music_list:
  press_times = len(df[df['has_press'] == True])
  all_times = len(df)
  print(f'押下率{press_times / all_times}')
```
